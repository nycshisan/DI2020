```
import matplotlib
# matplotlib.rcParams['text.usetex'] = True
from matplotlib import pyplot as plt 
import numpy as np
```

```
with open("TestCAResult.txt") as caResultFile:
    lines = caResultFile.readlines()
    lineNum = len(lines)
    assert lineNum % 2 == 0

    lengths = []
    memoriesBFShort = []
    memoriesCAShort = []
    memoriesBFLong = []
    memoriesCALong = []
    readTimesBFShort = []
    readTimesCAShort = []
    readTimesBFLong = []
    readTimesCALong = []
    writeTimesBFShort = []
    writeTimesCAShort = []
    writeTimesBFLong = []
    writeTimesCALong = []

    for i in range(lineNum // 2):
        lineShort = lines[i * 2].split()
        lineLong = lines[i * 2 + 1].split()
        lengths.append(int(lineShort[0]))

        memoriesBFShort.append(int(lineShort[2]))
        memoriesCAShort.append(int(lineShort[3]))
        readTimesBFShort.append(int(lineShort[4]))
        readTimesCAShort.append(int(lineShort[5]))
        writeTimesBFShort.append(int(lineShort[6]))
        writeTimesCAShort.append(int(lineShort[7]))
        
        memoriesBFLong.append(int(lineLong[2]))
        memoriesCALong.append(int(lineLong[3]))
        readTimesBFLong.append(int(lineLong[4]))
        readTimesCALong.append(int(lineLong[5]))
        writeTimesBFLong.append(int(lineLong[6]))
        writeTimesCALong.append(int(lineLong[7]))

    lengths = np.log(np.array(lengths))
    memoriesBFShort = np.log(np.array(memoriesBFShort))
    memoriesBFLong = np.log(np.array(memoriesBFLong))
    memoriesCAShort = np.log(np.array(memoriesCAShort))
    memoriesCALong = np.log(np.array(memoriesCALong))
```

```
plt.plot(lengths, memoriesBFShort, label='Brute Force')
plt.plot(lengths, memoriesCAShort, label='Succinct (Short Words)')
plt.plot(lengths, memoriesCALong, label='Succinct (Long Words)')
plt.title("Memory Cost for Arrays")
plt.xlabel("lg Length")
plt.ylabel("lg Memory (byte)")
plt.legend(loc='left right')
plt.show()

plt.plot(lengths, readTimesBFShort, label='Brute Force (Short Words)')
plt.plot(lengths, readTimesBFLong, label='Brute Force (Long Words)')
plt.plot(lengths, readTimesCAShort, label='Succinct (Short Words)')
plt.plot(lengths, readTimesCALong, label='Succinct (Long Words)')
plt.title("Reading Time for Arrays")
plt.xlabel("lg Length")
plt.ylabel("Avg. Time (ns)")
plt.legend(loc='left right')
plt.show()

plt.plot(lengths, writeTimesBFShort, label='Brute Force (Short Words)')
plt.plot(lengths, writeTimesBFLong, label='Brute Force (Long Words)')
plt.plot(lengths, writeTimesCAShort, label='Succinct (Short Words)')
plt.plot(lengths, writeTimesCALong, label='Succinct (Long Words)')
plt.title("Writing Time for Arrays")
plt.xlabel("lg Length")
plt.ylabel("Avg. Time (ns)")
plt.legend(loc='left right')
plt.show()
```

```
with open("TestBVResult.txt") as bvResultFile:
    lines = bvResultFile.readlines()
    lengths = []
    indexTimesBF = []
    indexTimesCA = []
    memoriesBF = []
    memoriesCA = []
    rankTimesBF = []
    rankTimesCA = []
    selectTimesBF = []
    selectTimesCA = []
    for line in lines:
        results = line.split()
        lengths.append(int(results[0]))
        indexTimesBF.append(int(results[1]))
        indexTimesCA.append(int(results[2]))
        memoriesBF.append(int(results[3]))
        memoriesCA.append(int(results[4]))
        rankTimesBF.append(int(results[5]))
        rankTimesCA.append(int(results[6]))
        selectTimesBF.append(int(results[7]))
        selectTimesCA.append(int(results[8]))
lengths = np.log(np.array(lengths))
indexTimesBF = np.log(np.array(indexTimesBF))
indexTimesCA = np.log(np.array(indexTimesCA))
memoriesBF = np.log(np.array(memoriesBF))
memoriesCA = np.log(np.array(memoriesCA))
rankTimesBF = np.array(rankTimesBF)
rankTimesCA = np.array(rankTimesCA)
selectTimesBF = np.array(selectTimesBF)
selectTimesCA = np.array(selectTimesCA)
```

```
plt.plot(lengths, indexTimesBF, label='Brute Force')
plt.plot(lengths, indexTimesCA, label='Succinct')
plt.title("Index Time for Bit Vectors")
plt.xlabel("lg Length")
plt.ylabel("lg Time (us)")
plt.legend(loc='left right')
plt.show()

plt.plot(lengths, memoriesBF, label='Brute Force')
plt.plot(lengths, memoriesCA, label='Succinct')
plt.title("Memory Cost for Bit Vectors")
plt.xlabel("lg Length")
plt.ylabel("lg Memory (byte)")
plt.legend(loc='left right')
plt.show()

plt.plot(lengths, rankTimesBF, label='Brute Force')
plt.plot(lengths, rankTimesCA, label='Succinct')
plt.title("Rank Time for Bit Vectors")
plt.xlabel("lg Length")
plt.ylabel("Avg. Time (ns)")
plt.legend(loc='right')
plt.show()

plt.plot(lengths, selectTimesBF, label='Brute Force')
plt.plot(lengths, selectTimesCA, label='Succinct')
plt.title("Select Time for Bit Vectors")
plt.xlabel("lg Length")
plt.ylabel("Avg. Time (ns)")
plt.legend(loc='right')
plt.show()
```
